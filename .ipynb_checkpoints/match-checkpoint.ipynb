```python
import pandas as pd
import numpy as np
```

# 1. Load datasets and Simmilarity matrices

```python
# load databases
df_publications = pd.read_parquet('/export/usuarios_ml4ds/mbalairon/github/recommendation_system/dataset/publications_topics.parquet')
df_eic = pd.read_parquet('/export/usuarios_ml4ds/mbalairon/github/recommendation_system/dataset/eic_topics.parquet')
df_horizon = pd.read_parquet('/export/usuarios_ml4ds/mbalairon/github/recommendation_system/dataset/horizon_topics.parquet')
df_publications_researchers = pd.read_parquet('/export/usuarios_ml4ds/mbalairon/github/data_ingest/UC3M ResearchPortal/Outputs/researchers_publications.parquet')
df_researchers = pd.read_parquet('/export/usuarios_ml4ds/mbalairon/github/data_ingest/UC3M ResearchPortal/Outputs/researchers.parquet')

# join together al the calls (igual que cuando creamos las matrices de similitud)
df_eic['Call'] = df_eic['id']
df_calls = pd.concat([df_horizon[['Call', 'tpc_decomposition']], df_eic[['Call', 'tpc_decomposition']]], axis=0).reset_index(drop=True)
```

```python
# load simmilarity matrixs
sim_matrix_embeddings = np.load('/export/usuarios_ml4ds/mbalairon/github/recommendation_system/simmilarity_matrices/cosine_sim_matrix_we_doc2vec.npy')
sim_matrix_BERT = np.load('/export/usuarios_ml4ds/mbalairon/github/recommendation_system/simmilarity_matrices/cosine_sim_matrix_we_BERT.npy')
sim_matrix_bhattacharyya = np.load('/export/usuarios_ml4ds/mbalairon/github/recommendation_system/simmilarity_matrices/bhattacharyya_sim_matrix.npy')
```

# 2. Get simmilarities for researchers

```python
def get_simmilarity_publication(sim_matrix, df_publications):
    '''
    Function for getting a dataset with the id of each of the publications and the simmilarity with each of the calls
    
    sim_matrix -> simmilarity matrix 
    df_publication -> df containing the publications 
    '''
    df = pd.DataFrame(columns=['actID', 'simmilarity'])
    for i in range(sim_matrix.shape[0]):
        df.loc[i, 'actID'] = df_publications.loc[i, 'actID']
        df.loc[i, 'simmilarity'] = sim_matrix[i]
    
    return df
        
```

```python
df_sim_pub_embeddings =  get_simmilarity_publication(sim_matrix_embeddings, df_publications)
df_sim_pub_embeddings_BERT =  get_simmilarity_publication(sim_matrix_BERT, df_publications)
df_sim_pub_bhattacharyya =  get_simmilarity_publication(sim_matrix_bhattacharyya, df_publications)
```

```python
df_pub_res_sim_embeddings = pd.merge(df_publications_researchers, df_sim_pub_embeddings, on='actID', how='inner')
df_pub_res_sim_embeddings = df_pub_res_sim_embeddings.groupby('invID')['simmilarity'].sum().reset_index()

df_pub_res_sim_BERT = pd.merge(df_publications_researchers, df_sim_pub_embeddings_BERT, on='actID', how='inner')
df_pub_res_sim_BERT = df_pub_res_sim_BERT.groupby('invID')['simmilarity'].sum().reset_index()

df_pub_res_sim_bhattacharyya = pd.merge(df_publications_researchers, df_sim_pub_bhattacharyya, on='actID', how='inner')
df_pub_res_sim_bhattacharyya = df_pub_res_sim_bhattacharyya.groupby('invID')['simmilarity'].sum().reset_index()
```

# 3. Create simmilarity matrices for researchers and calls

```python
def get_sim_matrix(df_calls, df_researchers):
    '''
    Function for computing the simmilarity matrix for the researchers and the calls
    
    df_calls -> dataframe with all the avaliable calls
    df_researchers -> Dataframe containing the researchers and the simmilarity with each of the publications
    '''
    
    keys_calls = df_calls['Call'].tolist()
    keys_res = df_researchers['invID'].tolist()
    sim = df_researchers['simmilarity'].tolist()
    
    df = pd.DataFrame(sim, columns=keys_calls, index=keys_res)
    
    return df

```

```python
simmilarity_embeddings = get_sim_matrix(df_calls, df_pub_res_sim_embeddings)
simmilarity_BERT = get_sim_matrix(df_calls, df_pub_res_sim_BERT)
simmilarity_battacharyya = get_sim_matrix(df_calls, df_pub_res_sim_bhattacharyya)
```

```python
simmilarity_embeddings.to_parquet('/export/usuarios_ml4ds/mbalairon/github/recommendation_system/simmilarity_matrices_researchers/simmilarity_embeddings.parquet', index=False)
simmilarity_BERT.to_parquet('/export/usuarios_ml4ds/mbalairon/github/recommendation_system/simmilarity_matrices_researchers/simmilarity_BERT.parquet', index=False)
simmilarity_battacharyya.to_parquet('/export/usuarios_ml4ds/mbalairon/github/recommendation_system/simmilarity_matrices_researchers/simmilarity_bhattacharyya.parquet', index=False)
```

# 4. perform the match 

```python
def match(simmilarities, call, n, df_researchers):
    '''
    Function for obtaining the ranking of researchers given a call 
    
    simmilarities -> df with all the simmilarities between researchers and calls
    call -> Call of interest
    n -> Number of researchers we are interested in 
    df_researchers -> Dataframe with the information about the researchers
    '''
    
    ranking = simmilarities[call].sort_values(ascending=False)
    id_researchers = ranking.nlargest(n).index
    id_researchers = pd.DataFrame(data=id_researchers, columns=['invID'])
    df_ranking_researchers = pd.merge(df_researchers, id_researchers, on='invID', how='inner')
    
    return df_ranking_researchers
```

```python
ranking = match(simmilarity_embeddings, '1', 3, df_researchers)
ranking
```

```python
ranking = match(simmilarity_BERT, '1', 3, df_researchers)
ranking
```

```python
ranking = match(simmilarity_battacharyya, '1', 3, df_researchers)
ranking
```
